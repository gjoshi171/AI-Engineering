### Preprocess title and features

```python
def preprocessed_description(row):
    return f"{row["title"]} {' '.join(row['features'])}"
```

```python
def extract_first_large_image(row):
    return row["images"][0].get("large", "")
```

```python
df_items["preprocessed_description"] = df_items.apply(preprocessed_description, axis=1)
df_items["image"] = df_items.apply(extract_first_large_image, axis=1)
```

```python
df_data_to_embed = df_sample[["preprocessed_description", "image", "rating_number", "price", "average_rating", "parent_asin"]]
```

```python
data_to_embed = df_data_to_embed.to_dict(orient="records")
```

### Define embedding function

```python
response = openai.embeddings.create(
    input ="Random text", 
    model="text-embedding-3-small"
)
```

```python
response.data[0].embedding
```

```python
len(response.data[0].embedding)
```

```python
def get_embedding(text, model = "text-embedding-3-small"):
    response = openai.embeddings.create(
        input =text, 
        model=model
    )
    return response.data[0].embedding  
```

```python
get_embedding("Hi")
```

### Create Qdrant Collection

```python
qdrant_client = QdrantClient(url = "http://localhost:6333")
```

```python
qdrant_client.create_collection(
    collection_name="Amazon-items-collection-01",
    vectors_config = VectorParams(size=1536, distance= Distance.COSINE)
)
```

### Embed Data

#### Pintstruct example

```python
pointstruct = PointStruct(
    id=0,
    vector= get_embedding("Hello, world!"),
    payload={
        "text": "Hello, world!",
        "model": "text-embedding-3-small"
    }
)
```

```python
# PointStruct represents a single “point”
pointstruct
```

### store data_to_embed in qdrant

```python
pointstructs = []
for i, data in enumerate(data_to_embed):
    embedding = get_embedding(data["preprocessed_description"])
    pointstructs.append(
        PointStruct(
            id = i,
            vector = embedding,
            payload = data
    ))
```

```python
len(pointstructs)
```

```python
# Update if the item exists, otherwise insert it
qdrant_client.upsert(
    collection_name = "Amazon-items-collection-01",
    points = pointstructs
)
```

### Define a function for data retrieval

```python
def retrieve_data(query, k=5):
    query_embedding = get_embedding(query)
    results = qdrant_client.query_points(
        collection_name = "Amazon-items-collection-01",
        query = query_embedding,
        limit = k
    )
    return results
```

```python
retrieve_data("Do you have a usb connectable fan for hot summers", k=10).points
```
